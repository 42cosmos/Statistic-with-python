```python
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from random import random
from scipy.special import comb
import scipy as sp

sns.set(style='ticks', color_codes = True)


import warnings
warnings.filterwarnings(action='ignore')
```

```python
df = pd.read_csv(os.path.join('data', 'StudentsPerformance.csv'))
df.shape
```

## 공분산, 상관계수

```python
np.cov(df[['math score', 'reading score']])
```

```python
np.cov(df[['math score', 'reading score']].T)
```

```python
df[['math score', 'reading score']].cov()
```

```python
df[['math score', 'reading score', 'writing score']].corr()
```

```python
np.corrcoef(df[['math score', 'reading score', 'writing score']].T)
```

## 선형그래프

```python
sns.pairplot(
    df[['math score', 'reading score', 'writing score']])
```

```python
sns.scatterplot(x = 'writing score', y = 'reading score', data = df)
```

```python
sns.relplot(x = 'writing score', y = 'reading score', data = df,) 
# Linear Regression ?
```

```python
# 보기 어렵따
sns.relplot(x = 'writing score', y = 'reading score', hue = 'race/ethnicity', data = df,) 
```

```python
sns.relplot(x = 'writing score', y = 'reading score', hue = 'gender', data = df,
           col = 'test preparation course', row = 'race/ethnicity') 
```

```python
sns.relplot(x = 'writing score', y = 'reading score', hue = 'writing score', data = df)
```

```python
# 성별의 차이는 이 그래프로는 모르겠다
sns.relplot(x = 'writing score', y = 'reading score', size = 'writing score', data = df, style='gender')
```

```python
# Linear Regression 값도 함께 보여줌
sns.jointplot(x='writing score', y='reading score', data = df, kind='reg')
```

```python
sns.jointplot(x='writing score', y='reading score', data = df, kind='hex')
```

```python
# kernel density estimation~ 등고선형
sns.jointplot(x='writing score', y='reading score', data = df, kind='kde')
```
