## 확률
- 데이터의 분포를 보는 하나의 지표 
- 다양한 결합확률을 보며 가설을 설계함
    - 결합확률을 보며 데이터를 뜯으며 가설을 세우는 게 중요
- 통계적 독립을 따로 사용치는 않음

# 결합확률 & 조건부 확률
실제 분석 시는 결합, 조건부확률이 많이 쓰인다.

```python
for col_name in object_cols:
    print(f" --------------- {col_name} --------------- ")
    print(df[col_name].value_counts(normalize=True))
```

```python
df.groupby(['parental level of education', 'gender']).count()
```

```python
# 부모 학위에 따른 성별의 확률
df.groupby(['parental level of education', 'gender'])['lunch'].count().unstack()
```

```python
# 조건부 확률 아님
df.groupby(['parental level of education', 'gender'])['lunch'].count().unstack() / df['lunch'].count()
```

```python
# 조건부 확률
# female(male) 이면서 부모가 어떤 학위를 가지고 있을 확률 
df.groupby(['parental level of education', 'gender'])['lunch'].count().unstack() / df.groupby(['gender'])['lunch'].count()
```

```python
df.groupby(['gender'])['lunch'].count()
```

```python
# 카테고리 플롯
sns.catplot('parental level of education', col='gender', data=df, kind='count', height=10, aspect=.8)
```

```python
# 부모에 학력에 따른 선행학습률?
sns.catplot('test preparation course', 
            col='parental level of education', data=df, 
            kind='count', height=10, aspect=.8, col_wrap=2)
```

```python
# 연속형 값을 한번에 보는 방법
# 산점도 시각화 및 각 값의 분포를 확인
sns.pairplot(df[numeric_cols])
```

```python
g = sns.PairGrid(df[numeric_cols])
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels= 6)
# reading, writing score > 차이가 많이 난다!
# 넓게 분포가 되어있다 
```

```python
# 점수와 (연속) 성별 (카테고리) 를 함께 보는 방법
sns.pairplot(df[numeric_cols.tolist() + ['gender']], hue='gender')
```

```python
# 여러 개를 함께 봄
# 젠더를 기준으로 수학점수만 보는 방법
g = sns.FacetGrid(df, col='gender')
g.map(plt.hist, 'math score')
```

```python
# 연속형 값과 이산형 값을 함께 볼 수 있음
g = sns.FacetGrid(df, col='parental level of education')
g.map(plt.hist, 'math score')
```

```python
# 어디에 값이 집중되어 있는가? 
sns.jointplot(x='math score', y='reading score', data=df, kind='kde')
```

```python
sns.jointplot(x='math score', y='reading score', data=df, kind='hex')
```
